```python
import pandas as pd
from glob import glob
import bioframe as bf
import numpy as np
import bioframe.vis
import matplotlib.pyplot as plt
```

```python
processed_path = './processed_files/'
```

```python
matrices_paths = glob(processed_path+'E*')
```

```python
k = 2
matrix_df = pd.read_csv(matrices_paths[k], sep='\t')
```

```python
def create_reads(matrix_df, cell_types=4):
    reads_df = matrix_df.iloc[:,0:3].copy()
    for i in range(cell_types):
        reads_df['cell_'+str(i)] = np.round(np.random.rand(len(reads_df))*100, 0, ).astype(int)
        
    return reads_df
```

```python
reads_df = create_reads(matrix_df)
```

```python
reads_df
```

```python
for path in matrices_paths:
    matrix_df = pd.read_csv(path, sep='\t')
    reads_df = create_reads(matrix_df)
    out_path = path+'.dread'
    reads_df.to_csv(out_path, sep='\t', index=False)
```
